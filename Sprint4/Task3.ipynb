Описание задачи: Создать базовый класс Student в C#, который будет представлять информацию о студентах. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма. Требования к базовому классу Student: • Атрибуты: Имя (Name), Возраст (Age), Курс (Course). • Методы: o GetInfo(): метод для получения информации о студенте в виде строки. o Study(): метод для вывода сообщения о том, что студент учится. o TakeExam(): метод для вывода сообщения о сдаче экзамена. Требования к производным классам:

Студент бакалавриата (BachelorStudent): Должен содержать дополнительные атрибуты, такие как Специальность (Specialty). Метод Study() должен быть переопределен для добавления информации о специальности при изучении предмета. Студент магистратуры (MasterStudent): Должен содержать дополнительные атрибуты, такие как Научный руководитель (ScientificAdvisor). Метод TakeExam() должен быть переопределен для добавления информации о научном руководителе при сдаче экзамена. Аспирант (PhDStudent) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Тема диссертации (DissertationTopic). Метод Study() должен быть переопределен для добавления информации о теме диссертации при изучении предмета. Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```
using System;
using System.Collections.Generic;

public class Student
{
    public string Name { get; set; }
    public int Age { get; set; }
    public int Course { get; set; }
    public int Score { get; set; }

    public Student(string name, int age, int course, int score)
    {
        Name = name;
        Age = age;
        Course = course;
        Score = score;
    }
    public virtual void GetInfo()
    {
        Console.WriteLine($"Информация о студенте: Имя: {Name}; Возраст: {Age};  Курс: {Course}");
    }
    public virtual void Study(int course)
    {
        Console.WriteLine($"Студент учится на {Course} курсе");
    }
    public virtual void Study(string specialty)
    {
        Console.WriteLine($"Студент учится на направлении: {specialty}");
    }
    public virtual void TakeExam()
    {
        if (Score > 64)
        {
            Console.WriteLine("Студент сдал экзамен");
        }
        else
        {
            Console.WriteLine("Студент отправлен на пересдачу");
        }
    }
}

public interface IBachStudent //интерфейс для явной реализации
{
    void Napravlenie();
    void Kurs();
}

public class Bakalavr
{
    private readonly IBachStudent _student;

    public Bakalavr(IBachStudent student)
    {
        _student = student;
    }
    public virtual void Kurs() // использование зависимости
    {
        _student.Kurs();
    }
    public virtual void Napravlenie() // использование зависимости
    {
        _student.Napravlenie();
    }
}

public class BachelorStudent : Student, IBachStudent
{
    private string _specialty;
    public string Specialty
    {
        get { return _specialty; }
        set { _specialty = value; }
    }

    public BachelorStudent(string name, int age, int course, int score, string specialty) : base(name, age, course, score)
    {
        Specialty = specialty;
    }
    void IBachStudent.Kurs() // явная реализация интерфейса
    {
        Console.WriteLine($"Студент-бакалавр учится на курс: {Course}"); // внедрение зависимости
    }
    void IBachStudent.Napravlenie() // явная реализация интерфейса
    {
        Console.WriteLine($"Также студент-бакалавр учится на направлении: {Specialty}"); // внедрение зависимости
    }

}

public class MasterStudent : BachelorStudent, IBachStudent
{
    private string _scientificadvisor;
    public string Scientificadvisor
    {
        get { return _scientificadvisor; }
        set { _scientificadvisor = value; }
    }
    public MasterStudent(string name, int age, int score, int course, string specialty, string scientificadvisor) : base(name, age, course, score, specialty)
    {
        Scientificadvisor = scientificadvisor;
    }

    public virtual void Napravlenie() // явная реализация интерфейса в классе 
    {
        Console.WriteLine($"Также студент учится на направлении: {Specialty}"); // внедрение зависимости
    }

    public virtual void Kurs() // явная реализация интерфейса в классе 
    {
        Console.WriteLine($"Также студент учится на {Course} курсе"); // внедрение зависимости
    }

    public override void TakeExam()
    {
        if (Score > 64)
        {
            Console.WriteLine($"Студент сдал экзамен научному руководителю /ФИО/: {Scientificadvisor}");
        }
        else
        {
            Console.WriteLine($"Студент сдавал экзамен научному руководителю /ФИО/: {Scientificadvisor}, но был отправлен на пересдачу");
        }
    }
}

public class PhDStudent : BachelorStudent
{
    private string _dissertationtopic;
    public string DissertationTopic
    {
        get { return _dissertationtopic; }
        set { _dissertationtopic = value; }
    }
    public PhDStudent(string name, int age, int course, int score, string dissertationtopic) : base(name, age, course, score, dissertationtopic)
    {
        DissertationTopic = dissertationtopic;
    }
    public void Study(int course, string dissertationtopic)
    {
        Console.WriteLine($"Студент учится на {Course} курсе и защищает диссертацию на тему: {DissertationTopic}");
    }
}

public class Students<T> where T : Student
{
    private List<T> _students = new List<T>();

    public void Add(T studenttype)
    {
        _students.Add(studenttype);
    }

    public void Remove(T studenttype)
    {
        _students.Remove(studenttype);
    }

    public void DisplayStudents()
    {
        foreach (var student in _students)
        {
            student.GetInfo();
        }
    }
}

IBachStudent backstud2 = new BachelorStudent("Данил", 21, 3, 80, "Прикладная математика и информатика");
Bakalavr kursnaprbak = new Bakalavr(backstud2);
Student defstudent = new Student("Вадим", 19, 2, 63);
BachelorStudent backstud = new BachelorStudent("Данил", 21, 3, 80, "Прикладная математика и информатика");
MasterStudent magistrstud = new MasterStudent("Кирилл", 22, 93, 4, "Прикладная математика и информатика", "Попов И.И.");
PhDStudent aspirstud = new PhDStudent("Ярослав", 23, 4, 90, "Теоретические основы информатики");

Console.WriteLine("Класс: Student-----------------------------------------");

defstudent.GetInfo();
defstudent.Study(defstudent.Course);
Console.WriteLine($"У студента: {defstudent.Score} баллов");
defstudent.TakeExam();

Console.WriteLine("-------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: BachelorStudent---------------------------------------------------------");

backstud.GetInfo();
kursnaprbak.Kurs();
kursnaprbak.Napravlenie();
Console.WriteLine($"У студента: {backstud.Score} баллов");
backstud.TakeExam();

Console.WriteLine("-------------------------------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: MasterStudent-----------------------------------------------------------");

magistrstud.GetInfo();
Console.WriteLine($"У студента: {magistrstud.Score} баллов");
magistrstud.TakeExam();
magistrstud.Napravlenie();

Console.WriteLine("-------------------------------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: PhDStudent--------------------------------------------------------------");

aspirstud.GetInfo();
aspirstud.Study(aspirstud.Course, aspirstud.DissertationTopic);
Console.WriteLine($"У студента: {aspirstud.Score} баллов");
aspirstud.TakeExam();

Console.WriteLine("-------------------------------------------------------------------------------");

Students<Student> students = new Students<Student>();
Console.WriteLine("Информация о бакалавре и аспиранте:");
students.Add(new BachelorStudent(backstud.Name, backstud.Age, backstud.Course, backstud.Score, backstud.Specialty)); // так как метод GetInfo, то последние переменные не выводятся так как метод под них не предусмотрен
students.Add(new PhDStudent(aspirstud.Name, aspirstud.Age, aspirstud.Course, aspirstud.Score, aspirstud.DissertationTopic));

students.DisplayStudents();

Console.WriteLine("-------------------------------------------------------------------------------");
```
