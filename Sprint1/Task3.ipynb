### Вариант задания 2


<h2 style="color:DodgerBlue">Описание проекта:</h2>

Описание задачи: Создать базовый класс Student в C#, который будет представлять информацию о студентах. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма. Требования к базовому классу Student: • Атрибуты: Имя (Name), Возраст (Age), Курс (Course). • Методы: o GetInfo(): метод для получения информации о студенте в виде строки. o Study(): метод для вывода сообщения о том, что студент учится. o TakeExam(): метод для вывода сообщения о сдаче экзамена. Требования к производным классам:

Студент бакалавриата (BachelorStudent): Должен содержать дополнительные атрибуты, такие как Специальность (Specialty). Метод Study() должен быть переопределен для добавления информации о специальности при изучении предмета. Студент магистратуры (MasterStudent): Должен содержать дополнительные атрибуты, такие как Научный руководитель (ScientificAdvisor). Метод TakeExam() должен быть переопределен для добавления информации о научном руководителе при сдаче экзамена. Аспирант (PhDStudent) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Тема диссертации (DissertationTopic). Метод Study() должен быть переопределен для добавления информации о теме диссертации при изучении предмета.

#### Дополнительное задание
Добавьте к сущестующим классам конструктора классов с использованием гетторов и сетторов и реализуйте взаимодействие объектов между собой

<h2 style="color:DodgerBlue">Реализация:</h2>

```
class Student 
{
    public string Name
    { get; set; }
    public int Age
    { get; set; }
    public int Course
    { get; set; }
    public int Score
    { get; set; }

    public Student(string name, int age, int course, int score)
    {
        Name = name;
        Age = age;
        Course = course;
        Score = score;
    }
    public virtual void GetInfo()
    {
        Console.WriteLine("Информация о студенте: " + "Имя: " + Name + "; Возраст: " + Age + "; Курс: " + Course);
    }
    public virtual void Study()
    {
        Console.WriteLine("Студент учится на " + Course + " курсе");
    }
    public virtual void TakeExam()
    {
        if (Score > 64)
        {
            Console.WriteLine("Студент сдал экзамен");
        }
        else
        {
            Console.WriteLine("Студент отправлен на пересдачу");
        }
    }
}
class BachelorStudent : Student
{
    private string _specialty;
    public string Specialty
    {
        get { return _specialty; }
        set { _specialty = value; }
    }
    public BachelorStudent(string name, int age, int course, int score, string specialty) : base(name, age, course, score)
    {
        Specialty = specialty;
    }
    public void StudyBach()
    {
        Console.WriteLine($"Студент учится на {Course} курсе и на направлении: {Specialty}");
    }
}

class MasterStudent : Student
{
    private string _scientificadvisor;
    public string Scientificadvisor
    {
        get { return _scientificadvisor; }
        set { _scientificadvisor = value; }
    }
    public MasterStudent(string name, int age, int course, int score, string scientificadvisor) : base(name, age, course, score)
    {
        Scientificadvisor = scientificadvisor;
    }
    public override void TakeExam()
    {
        if (Score > 64)
        {
            Console.WriteLine($"Студент сдал экзамен научному руководителю /ФИО/: {Scientificadvisor}");
        }
        else
        {
            Console.WriteLine($"Студент сдавал экзамен научному руководителю /ФИО/: {Scientificadvisor}, но был отправлен на пересдачу");
        }
    }
}

class PhDStudent : Student
{
    private string _dissertationtopic;
    public string DissertationTopic
    {
        get { return _dissertationtopic; }
        set { _dissertationtopic = value; }
    }
    public PhDStudent(string name, int age, int course, int score, string dissertationtopic) : base(name, age, course, score)
    {
        DissertationTopic = dissertationtopic;
    }
    public void StudyPhD()
    {
        Console.WriteLine($"Студент учится на {Course} курсе и защищает диссертацию на тему: {DissertationTopic}");
    }
}

Student defstudent = new Student("Анна", 19, 2, 63);
BachelorStudent backstud = new BachelorStudent("Елена", 21, 3, 80, "Прикладная математика и информатика");
MasterStudent magistrstud = new MasterStudent("Михаил", 20, 3, 40, "Попов. Ю. А.");
PhDStudent aspirstud = new PhDStudent("Алексей", 23, 4, 90, "Теоретические основы информатики");
Console.WriteLine("Класс: Student-----------------------------------------");

defstudent.GetInfo();
defstudent.Study();
Console.WriteLine($"У студента: {defstudent.Score} баллов");
defstudent.TakeExam();

Console.WriteLine("-------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: BachelorStudent---------------------------------------------------------");

backstud.GetInfo();
backstud.StudyBach();
Console.WriteLine($"У студента: {backstud.Score} баллов");
backstud.TakeExam();

Console.WriteLine("-------------------------------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: MasterStudent-----------------------------------------------------------");

magistrstud.GetInfo();
magistrstud.Study();
Console.WriteLine($"У студента: {magistrstud.Score} баллов");
magistrstud.TakeExam();

Console.WriteLine("-------------------------------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: PhDStudent--------------------------------------------------------------");

aspirstud.GetInfo();
aspirstud.StudyPhD();
Console.WriteLine($"У студента: {aspirstud.Score} баллов");
aspirstud.TakeExam();

Console.WriteLine("-------------------------------------------------------------------------------");
```
