### Вариант задания №2


<h2 style="color:DodgerBlue">Описание проекта:</h2>

----
Создать базовый класс Student в C#, который будет представлять информацию о
студентах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.
Требования к базовому классу Student:
• Атрибуты: Имя (Name), Возраст (Age), Курс (Course).
• Методы:
o GetInfo(): метод для получения информации о студенте в виде строки.
o Study(): метод для вывода сообщения о том, что студент учится.
o TakeExam(): метод для вывода сообщения о сдаче экзамена.
Требования к производным классам:
1. Студент бакалавриата (BachelorStudent): Должен содержать дополнительные
атрибуты, такие как Специальность (Specialty). Метод Study() должен быть
переопределен для добавления информации о специальности при изучении
предмета.
2. Студент магистратуры (MasterStudent): Должен содержать дополнительные
атрибуты, такие как Научный руководитель (ScientificAdvisor).
Метод TakeExam() должен быть переопределен для добавления
информации о научном руководителе при сдаче экзамена.
3. Аспирант (PhDStudent) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Тема диссертации (DissertationTopic).
Метод Study() должен быть переопределен для добавления информации о
теме диссертации при изучении предмета.


<h2 style="color:DodgerBlue">Реализация:</h2>

```polyglot-notebook
class Student 
{
    private string name;
    private int age;
    private int course;
    public string Name
    {
        get { return name; }
        set { name = value; }
    }
    public int Age
    {
        get { return age; }
        set { age = value; }
    }
    public int Course
    {
        get { return course; }
        set { course = value; }
    }
    public virtual void GetInfo()
    {
        Console.WriteLine("Информация о студенте: " + "Имя: " + Name + "; Возраст: " + Age + "; Курс: " + Course);
    }
    public virtual void Study()
    {
        Console.WriteLine("Студент учится на " + Course + " курсе");
    }
    public virtual void TakeExam()
    {
        Console.WriteLine("Студент сдал экзамен");
    }
}
class BachelorStudent : Student
{
    private string specialty;
    public string Specialty
    {
        get { return specialty; }
        set { specialty = value; }
    }
    public override void Study()
    {
        Console.WriteLine("Студент учится на " + Course + " курсе и на направлении: " + Specialty);
    }
}

class MasterStudent : Student
{
    private string scientificadvisor;
    public string ScientificAdvisor
    {
        get { return scientificadvisor; }
        set { scientificadvisor = value; }
    }
    public override void TakeExam()
    {
        Console.WriteLine("Студент сдал экзамен научному руководителю /ФИО/: " + ScientificAdvisor);
    }
}

class DStudent : Student
{
    private string dissertationtopic;
    public string DissertationTopic
    {
        get { return dissertationtopic; }
        set { dissertationtopic = value; }
    }
    public override void Study()
    {
        Console.WriteLine("Студент учится на " + Course + " курсе и защищает диссертацию на тему: " + DissertationTopic);
    }
}

Student student_ = new Student();
BachelorStudent backstud = new BachelorStudent();
MasterStudent magistrstud = new MasterStudent();
DStudent aspirstud = new DStudent();
Console.WriteLine("Класс: Student-----------------------------------------");

student_.Name = "Ирина";
student_.Age = 19;
student_.Course = 2;
student_.GetInfo();

student_.Study();

student_.TakeExam();
Console.WriteLine("-------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: BachelorStudent---------------------------------------------------------");

backstud.Name = "Галина";
backstud.Age = 22;
backstud.Course = 4;
backstud.Specialty = "Информационные системы и технологии в геологии и нефтегазовой отрасли";
backstud.GetInfo();

backstud.Study();

backstud.TakeExam();
Console.WriteLine("-------------------------------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: MasterStudent-----------------------------------------------------------");

magistrstud.Name = "Анастасия";
magistrstud.Age = 26;
magistrstud.Course = 5;
magistrstud.ScientificAdvisor = "Иванова С.Е.";
magistrstud.GetInfo();

magistrstud.Study();

magistrstud.TakeExam();
Console.WriteLine("-------------------------------------------------------------------------------");
Console.WriteLine("");
Console.WriteLine("Класс: DStudent--------------------------------------------------------------");

aspirstud.Name = "Егор";
aspirstud.Age = 32;
aspirstud.Course = 3;
aspirstud.DissertationTopic = "Методы и системы защиты информации";
aspirstud.GetInfo();

aspirstud.Study();

aspirstud.TakeExam();
Console.WriteLine("-------------------------------------------------------------------------------");
```
